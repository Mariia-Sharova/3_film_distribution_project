```python
# Прежде, чем объединять датасеты, приведем в порядок столбец `puNumber` в `data`
# Далее объединим датафреймы по столбцу с общим названием (столбец puNumber), методом `merge()`

# Так как обнаружилась ошибка, не дающая нам сменить формат в столбце, сначала удалим эту строку
data.query('puNumber == "нет"')
```

# 2.  Предобработка данных

## Проверьте типы данных в датафрейме и преобразуйте там, где это необходимо.

```python
df['show_start_date'].head(10)
```

```python
# Переводим дату в формат datetime
df['show_start_date']=pd.to_datetime(df['show_start_date'], format='%Y.%m.%dT')
df['show_start_date']=df['show_start_date'].dt.round('1D')# округляю до дня

#Создаю для удобства отдельный столбец с годом премьеры
df['year']=df['show_start_date'].dt.year

df.head()
```

```python
df.info()
```

```python
# У нас есть уникальные значения в столбце рейтинга в виде процентов, которые не дают поменять нам формат на float
df['ratings'].value_counts().sort_index(ascending=False)
```

```python
df['ratings'] = df['ratings'].str.replace('%','')
df['ratings'] = df['ratings'].astype('float64') 
```

```python
df['ratings'].value_counts().sort_index(ascending=False)
```
